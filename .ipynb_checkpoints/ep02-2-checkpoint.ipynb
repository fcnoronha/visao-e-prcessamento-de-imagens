Esse Jupyter Notebook representa a Parte 2 desta tarefa, que consiste em realizar a **normalização e análise da variação das classes** do dataset augmented. Para isso, vamos criar um novo dataset, sendo este:

- `normalized`: consiste das imagens de `augmented` normalizadas utilizando a equalização de histogramas como função inicial de normalização.

Criar para cada um dos datasets: original-gray, augmented, normalized:

    - Protótipo médio de cada classe;
    - Histograma médio de cada classe;
    - Variância do histograma de cada classe.
    

Para isso, utilizaremos metodos do pacote `scikit-image` e `cv2`.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, img_as_float
from skimage import exposure

from skimage import io
import os 

from skimage.util import img_as_ubyte
import cv2

# caminhos para os datasets
GREY_PATH = 'datasets/original-gray/'
NORM_PATH = 'datasets/normalized/'
AUG_PATH = 'datasets/augmented/'
```

# Normalização das imagens

```python
# Normalização das imagens que estão no dataset Augumented
for path, dirs, files in os.walk(AUG_PATH):
    for file in files:
        image_path = os.path.join(path, file)
        img = io.imread(image_path)
        img_eq = exposure.equalize_hist(img)
        io.imsave(NORM_PATH + file, img_as_ubyte(img_eq), quality=100)
```

# Protótipo médio de cada classe

# Histograma médio de cada classe e variância do histograma de cada classe.

```python
num_row = 10
num_col = 3
fig, axes = plt.subplots(num_row, num_col, figsize=(4*num_col, 4*num_row))
    
CODE_PATHS = [GREY_PATH, AUG_PATH, NORM_PATH]

#Passa por todos os datasets
for CODE_PATH, p in zip(CODE_PATHS, range(0,10)):
    
    files = os.listdir(CODE_PATH)
    files.sort()
    classes = list(set([x.split("_")[0] for x in files]))
    
    # Cria lista de todas as classes
    for class_, i in zip(classes, range(0,10)):
        class_files = [x for x in files if class_ in x and '_col' not in x]

        N=len(class_files)
        hist_array = []

        # Itera sobre todas as classes
        for file in class_files:

            # Le imagem do dataset
            image_path = CODE_PATH + file
            img = io.imread(image_path)
            
            hist_array.append(np.histogram(img, bins=256)[0])

        average_hist = np.matrix(hist_array).mean(0)

        ax = axes[i, p]
        ax.bar(range(0,256), np.reshape(average_hist.tolist(), -1))

        ax.set_title("Histograma da " + class_ + "\n Dataset: " + CODE_PATH.split("/")[1] + "\n Com variância " + str(round(np.var(np.reshape(average_hist.tolist(), -1)))))

plt.tight_layout()
plt.show()
```
